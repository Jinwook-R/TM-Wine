```python
## 품종 1개 인거 제거하기
def del_row(df,k):
    lists=df['품종']
    count={}
    b=[]
    a=[]
    for i in lists:
        try: count[i] += 1
        except: count[i]=1
    for i in count.keys():
        if count[i]==k:
            b.append(i)
    for i in range(len(b)):
        a.append(df.index[df['품종']==b[i]][0])
    df = df.drop(index=a, axis=0)
    return df
```
